```python
#  Numerical column analysis 

#  Maximum and mimimum value mean, meadian or std deviation or variance

#  You can learn about their distribution.
# - Histogram.
#  While a histogram is made up of bars that touch each other,it shows counts of values in each range.

# - Distribution plot. 
#  While a density plot shows the proportion of values in each range, a density plot is a smooth curve 
# that shows the distribution of the data in a more continuous way

# Asymmetry 
# Skewness is a statistical measure of the asymmetry of a probability distribution. 
# It characterizes the extent to which the distribution of a set of values deviates from a normal distribution.
# If the skewness is between -0.5 & 0.5, the data are nearly symmetrical.
# If the skewness is between -1 & -0.5 (negative skewed) or between 0.5 & 1(positive skewed), the data are slightly skewed.
# If the skewness is lower than -1 (negative skewed) or greater than 1 (positive skewed), the data are extremely skewed.

# Quantile Analysis
# A quantile defines a particular part of a data set, 
# i.e. a quantile determines how many values in a distribution are above or below a certain limit.

# Analysis of extreme value
# - Box Plot
# - Violin plot
```

```python
# Numerical columns
# 'Age', 'SibSp', 'Parch', 'Fare',
```

```python
df["Age"].mean()

df["Age"].min()

df["Age"].max()

df["Age"].std()

df["Age"].var()

# Skewness between -0.5 and 0.5 is symmetrical.
df["Age"].skew()

# this can interpreted as the highest peak of curve will around centre of middle of the graph.

#  Plot is histogram
#  Histogram gives information about population distibution

### What  is the age group of most of the people are travelling on titanic ship?

**what is the distibution of numeric column and where does most of the values lies?**

plt.figure(figsize=(8,6)) # (8,6) --("width","height")
# Histogram plot
sns.histplot(df["Age"]) # sns.histplot(df["Age"],kde=True)

plt.title("Histogram plot for Age",y=1.02)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,6))
# Distribution plot
sns.displot(df["Age"],kind="kde") # or sns.kdeplot(df["Age"])
# kde stands for kernel density estimation

plt.title("Density plot for Age",y=1.02)
plt.xlabel("Age")
plt.ylabel("Probability")
plt.show()
# OptionError: "No such keys(s): 'mode.use_inf_as_null'"
# Error due to seaborn version 0.11.2 
# Solution upgrade to 0.12.2 solved the issue.

### Are there any extreme values of age, if so what are they?

df["Age"].quantile(0.10),df["Age"].quantile(0.25),df["Age"].quantile(0.50),df["Age"].quantile(0.75),df["Age"].quantile(0.90)

plt.figure(figsize=(8,6))
# Box plot
sns.boxplot(df["Age"],orient="h") # orient="h" for horizontal or orient="v" for vertical

plt.title("Box plot for Age",y=1.02)
plt.xlabel("Age")
# plt.ylabel("")
plt.show()

plt.figure(figsize=(8,6))
# Violin Plot
sns.violinplot(df["Age"],orient="h",color="tomato") # refer color guide at bottom
sns.swarmplot( df["Age"],size=3,orient="h")

plt.title("Violin plot for Age",y=1.02)
plt.xlabel("Age")
# plt.ylabel("")
plt.show()
```

**Age** Column has a symmetrical distribution with skew 0.32 which is between -0.5 and 0.5 . 
- Density curve for Age is centred around mean 29.69 with standard deviation of 14.52.
- Range of Age value is from minimum 0.40 yrs and maximum 80 yrs.
- Most of the people are failling under age group of 20-40 yrs.
- 10% of people are of age 14 and below.
- 25% of people are of age 20 and below.
- 50% of people are of age 28 and below.
- 75% of people are of age 38 and below.
- 90% of people are of age 50 and below.
- Age column has few extreme values toward higher age group 65 yrs and above.

Age column could be a good predictor for Survival and it's relation with other columns should be explored further.
